```python
import numpy as np

def get_items(field):
    wood = 0
    lumberyards = 0
    empty = 0
    for elem in field.flatten():
        if elem == "|":
            wood += 1
        elif elem == "#":
            lumberyards += 1
        else:
            empty += 1
    return empty, wood, lumberyards


field = []

with open("input.txt") as inp:
    for line in map(lambda x: x.rstrip(), inp.readlines()):
        field.append(list(line))

field = np.array(field)
start_field = field.copy()
start_items = get_items(start_field)
cache = {}


def find_loop():
    last_diff = 0
    minutes = 0
    while True:
        items = minute()
        minutes += 1
        if items in cache:
            # print("Current minute:", minutes, "last seen in:", cache[items])
            diff = minutes - cache[items]
            if not last_diff:
                last_diff = diff
            else:
                if last_diff == diff:
                    return diff, minutes
                else:
                    last_diff = diff

        cache[items] = minutes


def minute(iter_num=1):
    global field
    for it in range(iter_num):
        new_field = field.copy()
        for i in range(len(field)):
            for j in range(len(field[i])):
                near = list(
                    field[max(i - 1, 0):min(len(field), i + 2), max(j - 1, 0):min(len(field[i]), j + 2)].flatten())
                if field[i, j] == "." and near.count("|") >= 3:
                    new_field[i, j] = "|"
                elif field[i, j] == "|" and near.count("#") >= 3:
                    new_field[i, j] = "#"
                if field[i, j] == "#":
                    if not (near.count("#") >= 2 and near.count("|") >= 1):
                        new_field[i, j] = "."

        field = new_field
    return get_items(field)


def string_field(field):
    print("--vvv--")
    for line in field:
        print(''.join(line))
    print("--^^^--")


dx, loop_started = find_loop()

iter_left = 1_000_000_000 - loop_started
need_to_loop = iter_left % dx

minute(need_to_loop)

empty, wood, lumberyards = get_items(field)

print(wood, lumberyards)
print(wood * lumberyards)
```
