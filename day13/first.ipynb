```python
field = []
carts = {}
cart_num = 0

# 0 - Left
# 1 - Up
# 2 - Right
# 3 - Down

with open("input.txt") as inp:
    for row_num, line in enumerate(inp.readlines()):
        row = line.rstrip()
        field.append(list(row))
        for j in range(len(row)):
            if row[j] in "<>^v":
                carts[cart_num] = ((row_num, j), ("<^>v".index(row[j]), 0))
                cart_num += 1


def get_dir(counter, direct):
    if counter % 3 == 0:
        return counter + 1, (direct - 1 if direct != 0 else 3)
    if counter % 3 == 1:
        return counter + 1, direct
    if counter % 3 == 2:
        return counter + 1, (direct + 1 if direct != 3 else 0)


def tick(carts):
    crashed = set()
    carts = sort_carts(carts)
    for cart_num in carts:
        i, j = carts[cart_num][0]
        direction, counter = carts[cart_num][1]
        if direction == 0:
            if field[i][j - 1] == "/":
                direction = 3
            elif field[i][j - 1] == "\\":
                direction = 1
            elif field[i][j - 1] == "+":
                counter, direction = get_dir(counter, direction)
            j -= 1
        elif direction == 1:
            if field[i - 1][j] == "/":
                direction = 2
            elif field[i - 1][j] == "\\":
                direction = 0
            elif field[i - 1][j] == "+":
                counter, direction = get_dir(counter, direction)
            i -= 1
        elif direction == 2:
            if field[i][j + 1] == "/":
                direction = 1
            elif field[i][j + 1] == "\\":
                direction = 3
            if field[i][j + 1] == "+":
                counter, direction = get_dir(counter, direction)
            j += 1
        elif direction == 3:
            if field[i + 1][j] == "/":
                direction = 0
            elif field[i + 1][j] == "\\":
                direction = 2
            elif field[i + 1][j] == "+":
                counter, direction = get_dir(counter, direction)
            i += 1

        carts[cart_num] = ((i, j), (direction, counter))
        crashed.update(check_crash(carts))

    return carts, crashed


def print_field():
    for i in range(len(field)):
        for j in range(len(field[i])):
            for cart in carts:
                if (i, j) == carts[cart][0]:
                    print("&", end="")
                    break
            else:
                print(field[i][j], end="")
        print()


def check_crash(carts):
    to_remove = []
    for cart in carts:
        for sub_cart in carts:
            if cart != sub_cart:
                if cart not in to_remove and sub_cart not in to_remove and carts[cart][0] == carts[sub_cart][0]:
                    to_remove.append(cart)
                    to_remove.append(sub_cart)
    assert len(to_remove) == len(set(to_remove))
    return to_remove


def sort_carts(carts):
    sorted_carts = sorted(carts, key=lambda x: carts[x][0])
    return {i: carts[key] for i, key in enumerate(sorted_carts)}


def solve(carts):
    while True:
        carts, crashed = tick(carts)
        if len(crashed) != 0:
            return ','.join(map(str, carts[next(iter(crashed))][0][::-1]))


print(solve(carts))
```
